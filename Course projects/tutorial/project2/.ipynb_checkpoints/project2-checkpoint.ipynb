### imputation of missing value

```python
def add_missing_values(x_full, y_full):
    n_samples, n_features = x_full.shape
    
    # Add missing values in 75% of the lines
    missing_rate = 0.75
    n_missing_samples = int(n_samples * missing_rate)
    
    missing_samples = np. zeros(n_samples, dtype=bool)
    missing_samples[: n_missing_samples] = True
    
    np.random.shuffle(missing_samples)
    missing_features = np.random.randint(0, n_features, n_missing_samples)
    X_missing =x_full.copy()
    X_missing[missing_samples, missing_features] = np. nan
    Y_missing = y_full.copy()
    return X_missing, Y_missing
```

```python
X_missing, Y_missing= add_missing_values(x_full, y_full)
```

```python
X_missing
```

### Simple Imputer

```python
from sklearn.impute import SimpleImputer
```

```python
si = SimpleImputer(strategy='most_frequent').fit(X_missing)
```

```python
x_recovered = si.transform(X_missing)
```

```python
np.sum(x_recovered * x_full)/(norm(x_full,2)*norm(x_recovered,2))
```

```python
from sklearn.pipeline import make_pipeline , Pipeline
from sklearn.preprocessing import StandardScaler
```

```python
steps = [('simple imputer with mean',SimpleImputer(strategy='mean').fit(x_full)),('standard scaler',StandardScaler().fit(x_full))]
preprocess_pipeline =Pipeline(steps)
```

```python
x_recovered2 = preprocess_pipeline.transform(X_missing)
```

```python
x_recovered2.mean(axis=0)
```

```python
x_recovered2.var(axis=0)
```
