### PCA

```python
from sklearn.datasets import load_iris
import numpy as np
```

```python
X,y=load_iris(return_X_y = True)
```

```python
X.shape
```

```python
y.shape
```

```python
np.unique(y)
```

```python
from sklearn.decomposition import PCA
from numpy.linalg import norm
```

```python
pca = PCA(n_components=2).fit(X)
```

```python
pca
```

```python
Xt = pca.transform(X)
```

```python
Xt.shape
```

```python

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10,10))
plt.scatter(Xt[:,0],Xt[:,1],c=y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
polyf = PolynomialFeatures().fit(X)
```

```python
Xf = polyf.transform(X)
```

```python
Xf.shape
```

```python
pca1= PCA(n_components=0.96).fit(Xf)
Xtf = pca1.transform(Xf)
```

```python
Xtf.shape
```

```python
plt.figure(figsize=(10,10))
plt.scatter(Xtf[:,0],Xtf[:,1],c=y)

plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
```

### LDA

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
```

```python
lda = LDA(n_components = 2).fit(X,y)
```

```python
Xt_lda = lda.transform(X)
```

```python
plt.figure(figsize=(10,10))
plt.scatter(Xt_lda[:,0],Xt_lda[:,1],c=y)

plt.xlabel('LDA 1')
plt.ylabel('LDA 2')
```
