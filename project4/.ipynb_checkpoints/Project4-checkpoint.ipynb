### Neighborhood Component Analysis

```python
from sklearn.neighbors import NeighborhoodComponentsAnalysis as NCA
from sklearn.datasets import load_iris
import numpy as np
from sklearn.decomposition import PCA
from numpy.linalg import norm
```

```python
X,y=load_iris(return_X_y = True)
```

```python
nca = NCA(n_components=2).fit(X,y)
Xt_nca=nca.transform(X)
```

```python
plt.figure(figsize=(10,10))
plt.scatter(Xt_nca[:,0],Xt_nca[:,1],c=y)

plt.xlabel('NCA 1')
plt.ylabel('NCA 2')
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
X, y = load_breast_cancer(return_X_y = True)
```

```python
X.shape
```

```python
np.unique(y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.2)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier as KNN
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler,MaxAbsScaler,MinMaxScaler,Normalizer
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
```

```python
X.mean(axis=0)
```

```python
X.std(axis=0)
```

```python
pca = PCA(n_components=2).fit(X)
X_pca = pca.transform(X)
```

```python
plt.figure(figsize=(10,10))
plt.scatter(X_pca[:,0],X_pca[:,1],c=y,alpha=0.5)

```

```python
knn = KNN(n_neighbors=5).fit(X_train,y_train)
```

```python
y_pred = knn.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
acc=0
for i in range(y_pred.shape[0]):
    if y_pred[i] == y_test[i]:
        acc += 1
acc /= y_pred.shape[0]
print('accuracy is {}'.format(acc))
```

```python
steps = [('preprocess',MaxAbsScaler()),
         ('dimension reduction',PCA(n_components=5)),
         #('dimension reduction',LDA(n_components=1)),
        ('Classifier',KNN(n_neighbors=5)),]
pipeline = Pipeline(steps).fit(X_train,y_train)
```

```python
y_pred = pipeline.predict(X_test)
```

```python
acc=0
for i in range(y_pred.shape[0]):
    if y_pred[i] == y_test[i]:
        acc += 1
acc /= y_pred.shape[0]
print('accuracy is {}'.format(acc))
```

### ICA 

```python
from scipy.signal import sawtooth
```

```python
n_sample = 5000
t = np.linspace(0,8,n_sample)
```

```python
s1 = np.sin(2*t)
s2 = np.sign(np.sin(2*t))
s3 = sawtooth(2*t)
```

```python
plt.plot(t, s1)
plt.figure()
plt.plot(t, s2)
plt.figure()
plt.plot(t, s3)
```

```python
S = np.asarray([s1,s2,s3])
```

```python
S.shape
```

```python
S += np.random.normal(0,1,S.shape)
```

```python
plt.plot(t,S[0,:])
plt.plot(t,S[1,:])
plt.plot(t,S[2,:])
```

```python
A = np.array([[1,2,-1],[0,-1,-1],[1,1,3]])
X = np.dot(A,S)
```

```python
plt.plot(t,X[0,:])
plt.plot(t,X[1,:])
plt.plot(t,X[2,:])
```

```python
from sklearn.decomposition import FastICA
```

```python
ica = FastICA(n_components=3).fit(X.transpose())
```

```python
ica
```

```python
Xt_ica = ica.transform(X.transpose())
```

```python
Xt_ica
```
