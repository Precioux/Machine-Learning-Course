```python
from sklearn.datasets import fetch_california_housing
```

```python
dataset = fetch_california_housing()
x_full= dataset.data
y_full= dataset.target
```

```python
x_full.shape
```

```python
y_full.shape
```

```python
y_full
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

import numpy as np
```

```python
plt.figure(figsize=(10,10))
plt.scatter(x_full[:,4],x_full[:,1],c=y_full)
```

```python
x_full.mean(axis=0)
```

```python
x_full.var(axis=0)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
stcaler = StandardScaler().fit(x_full)
```

```python
x_all_stcaler=stcaler.transform(x_full)
```

```python
x_all_stcaler.mean(axis=0)
```

```python
x_all_stcaler.var(axis=0)
```

```python
plt.figure(figsize=(10,10))
plt.scatter(x_all_stcaler[:,4],x_all_stcaler[:,1],c=y_full)
```

```python
plt.hist(x_full[:,0],bins=100)
plt.figure()
plt.hist(x_all_stcaler[:,0],bins=100)
```

```python
plt.hist(x_full[:,1],bins=100)
plt.figure()
plt.hist(x_all_stcaler[:,1],bins=100)
```

```python
plt.hist(x_full[:,4],bins=100)
plt.figure()
plt.hist(x_all_stcaler[:,4],bins=100)
```
